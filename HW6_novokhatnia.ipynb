```python
def check_train_score(params, lgb_data, data, target, kf, num_rounds):
    roc_auc = []
    for train, val in kf.split(data,target):
        temp_lgb_train = lgb_data.subset(train)
        temp_lgb_val = lgb_data.subset(val)
        temp_model = lgb.train(params, temp_lgb_train, num_rounds, verbose_eval=num_rounds)
        roc_auc.append([roc_auc_score(target[train],temp_model.predict(data.loc[train])), roc_auc_score(target[val],temp_model.predict(data.loc[val]))])
    return np.mean(roc_auc, axis=0)
```

```python
parameters_3 = {
    #default
    'objective': 'binary',
    'learning_rate': 0.01,
    'num_threads': 4,
    "metric": 'auc',
    
    #regularization
    'colsample_bytree': 0.8,
    'subsample': 0.9,
    'subsample_freq': 1,
    'min_data_in_leaf': 40,
    
    'num_leaves': 64,
    
    #categorical features
    'cat_smooth': 10,
    'min_data_per_group': 50
}


result_lgbm3 = lgb.cv(parameters_3, lgb_train, n_rounds, folds=skf.split(X, y), early_stopping_rounds=50, verbose_eval=1, )
```

```python
print(max(result_lgbm3['auc-mean']), len(result_lgbm3['auc-mean']))
print(check_train_score(parameters_3, lgb_train, X, y, skf, len(result_lgbm3['auc-mean'])))
```

```python
model_lgm = lgb.train(parameters_3, lgb_train, len(result_lgbm3['auc-mean']), verbose_eval=len(result_lgbm3['auc-mean']))
```
